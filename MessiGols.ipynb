```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('LionelMessi.csv')
```

```python
dataset.head(5)
```

```python
dataset.columns
```

```python
renomear = {
    'Season': 'Temporada', 
    'Competition' : 'Competição', 
    'Matchday' : 'Matchday', 
    'Date' : 'Dia', 
    'Venue' : 'Local Partida', 
    'Club' : 'Clube',
    'Opponent' : 'Adversário', 
    'Result' : 'Resultado', 
    'Playing_Position' : 'Posição Jogada', 
    'Minute' : 'Minutos', 
    'At_score': 'Pontuação', 
    'Type' : 'Tipo de Gol',
    'Goal_assist' : 'Assistência'
}

dataset = dataset.rename(columns = renomear)
```

```python
df = dataset[['Temporada', 'Competição','Matchday', 'Dia', 'Local Partida', 'Clube','Adversário', 'Resultado',  'Posição Jogada', 'Minutos', 'Pontuação', 'Tipo de Gol', 'Assistência']]
```

```python
df.head(5)
```

```python
# Transformando Date para tipo datatime
df['Dia'] = pd.to_datetime(df['Dia'])
```

```python
df.info()
```

# Gols em Competições


```python
plt.figure(figsize=(15, 6))
plt.title('Analisando Gols em Competições')
plt.bar(df['Competição'].value_counts().index, df['Competição'].value_counts());
plt.xticks(rotation=45, ha='right');
```

# Tipo de Gols 


```python
plt.figure(figsize=(15, 6))
plt.title('Analisando Tipo de Gols')
plt.bar(df['Tipo de Gol'].value_counts().index, df['Tipo de Gol'].value_counts());
plt.xticks(rotation=45, ha='right');
```

# Top 10 Adversários - Gols Marcados

```python
top10_adv = pd.DataFrame(df['Adversário'].value_counts())
top10_adv = top10_adv.head(10)
top10_adv
```

```python
plt.figure(figsize=(18, 8))

names = top10_adv.index
valores = top10_adv.Adversário

plt.barh(names, valores)

barlist = plt.barh(names, valores)
barlist[0].set_color('#04931b')
barlist[1].set_color('#04931b')
barlist[2].set_color('#bb7013')
barlist[3].set_color('#bb7013')
barlist[4].set_color('#bb7013')
barlist[5].set_color('#bb7013')
barlist[6].set_color('#e33c0a')
barlist[7].set_color('#e33c0a')
barlist[8].set_color('#e33c0a')
barlist[9].set_color('#e33c0a')

plt.ylabel('Nome')
plt.title('Top 10 Adversários onde mais fez Gol', fontsize=14)
plt.show()
     
```

# Top 10 Assistentes do Messi



```python
maiores_ass = round(df['Assistência'].value_counts())
maiores_ass = maiores_ass.head(10)
maiores_ass
```

```python
plt.figure(figsize=(15, 6))
plt.title('Maiores Assistentes do Messi')
plt.pie(
    maiores_ass.head(9), 
    labels = maiores_ass.index[0:9],
    shadow=True,
    startangle=90,
    autopct='%1.1f%%'
);
```
